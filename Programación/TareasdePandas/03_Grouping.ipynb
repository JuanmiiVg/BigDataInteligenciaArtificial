### Ejercicio 2: Load drinks
https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv

```python
# Pista: pd.read_csv()
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
```

```python
assert drinks.shape[0] > 0
print('âœ“ Cargado')
```

### Ejercicio 4: Beer by continent
Which continent drinks most beer?

```python
# Pista: .groupby('continent').beer.mean()
beer_mean = drinks.groupby('continent').beer_servings.mean()
```

```python
assert beer_mean is not None
print('âœ“ OK')
```

### Ejercicio 5: Wine stats
Stats for wine by continent

```python
# Pista: .groupby('continent').wine.describe()
wine_stats = drinks.groupby('continent').wine_servings.describe()
```

```python
assert wine_stats is not None
print('âœ“ OK')
```

### Ejercicio 6: Mean all cols
Mean by continent for all

```python
# Pista: .groupby('continent').mean()
all_mean = drinks.groupby('continent').mean(numeric_only=True)
```

```python
assert all_mean is not None
print('âœ“ OK')
```

### Ejercicio 7: Median all cols
Median by continent

```python
# Pista: .groupby('continent').median()
all_median = drinks.groupby('continent').median(numeric_only=True)
```

```python
assert all_median is not None
print('âœ“ OK')
```

### Ejercicio 8: Spirit agg
Mean, min, max for spirits

```python
# Pista: .agg(['mean', 'min', 'max'])
spirit_agg = drinks['spirit_servings'].agg(['mean', 'min', 'max'])
```

```python
assert spirit_agg is not None
print('âœ“ OK')
```

### Ejercicio 9: Count by continent
How many countries per continent?

```python
# Pista: .groupby('continent').size()
countries = drinks.groupby('continent').size()
```

```python
assert countries is not None
print('âœ“ OK')
```

### Ejercicio 10: Total consumption
Total alcohol by continent

```python
# Pista: .groupby('continent').sum()
total = drinks.groupby('continent').sum()
```

```python
assert total is not None
print('âœ“ OK')
```

### Ejercicio 11: Custom agg
Multiple functions

```python
# Pista: .agg({'beer': 'mean', 'wine': 'sum'})
custom = drinks.agg({'beer_servings': 'mean', 'wine_servings': 'sum'})
```

```python
assert custom is not None
print('âœ“ OK')
```

### Ejercicio 12: Filter + group
Mean beer where > 100

```python
# Pista: .loc[drinks.beer > 100].groupby('continent').mean()
filtered = drinks.loc[drinks.beer_servings > 100].groupby('continent').mean(numeric_only=True)
```

```python
assert filtered is not None
print('âœ“ OK')
```
