### Ejercicio 2: Load data
https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

```python
# Pista: pd.read_csv()
baby_names = pd.read_csv(
    "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
)
```

```python
assert baby_names.shape[0] > 0
print('âœ“ Cargado')
```

### Ejercicio 3: Preview
First rows

```python
# Pista: .head()
baby_names.head()
```

```python
print('OK')
print('âœ“ OK')
```

### Ejercicio 4: Drop cols
Delete 'Unnamed: 0' and 'Id'

```python
# Pista: .drop(['Unnamed: 0', 'Id'], axis=1)
bn_clean = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
```

```python
assert bn_clean is not None
print('âœ“ Limpio')
```

### Ejercicio 5: Gender count
More male or female names?

```python
# Pista: .groupby('Gender').size()
gender_count = bn_clean.groupby('Gender').size()
```

```python
assert gender_count is not None
print('âœ“ OK')
```

### Ejercicio 6: Group by name
Groupby name

```python
# Pista: .groupby('Name')
by_name = bn_clean.groupby('Name')
```

```python
assert by_name is not None
print('âœ“ OK')
```

### Ejercicio 7: Unique names
How many different names?

```python
# Pista: .nunique() or .groupby().ngroups
num_names = bn_clean['Name'].nunique()
num_names
```

```python
assert num_names > 0
print('âœ“ OK')
```

### Ejercicio 8: Most common
Name with most occurrences

```python
# Pista: .groupby('Name').Count.sum().idxmax()
most_common = bn_clean.groupby('Name')['Count'].sum().idxmax()
most_common
```

```python
assert most_common is not None
print('âœ“ OK')
```

### Ejercicio 9: Least common
How many with least?

```python
# Pista: ...min()...
total_counts = bn_clean.groupby('Name')['Count'].sum()
least_count = (total_counts == total_counts.min()).sum()
least_count
```

```python
assert least_count > 0
print('âœ“ OK')
```

### Ejercicio 10: Median
Median name occurrence

```python
# Pista: .groupby('Name').Count.sum().median()
med = bn_clean.groupby('Name')['Count'].sum().median()
med
```

```python
assert med > 0
print('âœ“ OK')
```

### Ejercicio 11: Std
Standard deviation

```python
# Pista: .std()
std_val = bn_clean['Count'].std()
```

```python
assert std_val > 0
print('âœ“ OK')
```

### Ejercicio 12: Describe
Summary stats

```python
# Pista: .describe()
summary = bn_clean.describe()
```

```python
assert summary is not None
print('âœ“ OK')
```

### Ejercicio 13: Top 10
Top 10 names

```python
# Pista: .groupby('Name').Count.sum().nlargest(10)
top10 = bn_clean.groupby('Name')['Count'].sum().nlargest(10)
```

```python
assert len(top10) == 10
print('âœ“ OK')
```

### Ejercicio 14: Percentiles
25th, 50th, 75th

```python
# Pista: .quantile([.25, .5, .75])
perc = bn_clean['Count'].quantile([0.25, 0.5, 0.75])
```

```python
assert perc is not None
print('âœ“ OK')
```

### Ejercicio 15: Trend
Names trend over time

```python
# Pista: .groupby(['Year', 'Name']).Count.sum()
trend = bn_clean.groupby(['Year', 'Name'])['Count'].sum()
```

```python
assert trend is not None
print('âœ“ OK')
```
