### Ejercicio 2: Load cars1
https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv

```python
# Pista: pd.read_csv()
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
```

```python
assert cars1 is not None
print('âœ“ Cargado')
```

### Ejercicio 3: Load cars2
https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv

```python
# Pista: pd.read_csv()
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")
```

```python
assert cars2 is not None
print('âœ“ Cargado')
```

### Ejercicio 4: Clean cars1
Remove unnamed columns

```python
# Pista: .loc[:, 'mpg':'car']
cars1_clean = cars1.loc[:, "mpg":"car"]
```

```python
assert cars1_clean is not None
print('âœ“ Limpio')
```

### Ejercicio 5: Check shape
Shape of each df

```python
# Pista: .shape
print(cars1_clean.shape, cars2.shape)
```

```python
print(f'{cars1_clean.shape}, {cars2.shape}')
print('âœ“ Revisado')
```

### Ejercicio 6: Concat
Join cars1 and cars2

```python
# Pista: pd.concat([cars1, cars2])
cars = pd.concat([cars1_clean, cars2])
```

```python
assert cars is not None
print('âœ“ Concatenado')
```

### Ejercicio 7: Create owners
Random integers 15000-73000

```python
# Pista: np.random.randint(15000, 73001, size=len(cars))
owners = np.random.randint(15000, 73001, size=len(cars))
```

```python
assert len(owners) == len(cars)
print('âœ“ Creado')
```

### Ejercicio 8: Add column
Add owners to cars

```python
# Pista: cars['owners'] = owners
cars_owners = cars.copy()
cars_owners["owners"] = owners
```

```python
assert 'owners' in cars_owners.columns
print('âœ“ AÃ±adido')
```

### Ejercicio 9: Merge condition
pd.merge on index

```python
# Pista: pd.merge(cars1, cars2, left_index=True, right_index=True)
merged = pd.merge(cars1_clean, cars2, left_index=True, right_index=True)
```

```python
assert merged is not None
print('âœ“ OK')
```

### Ejercicio 10: Inner join
Inner merge

```python
# Pista: pd.merge(..., how='inner')
inner = pd.merge(cars1_clean, cars2, how="inner", left_index=True, right_index=True)
```

```python
assert inner is not None
print('âœ“ OK')
```

### Ejercicio 11: Left join
Left merge

```python
# Pista: pd.merge(..., how='left')
left = pd.merge(cars1, cars2, how='left')
```

```python
assert left is not None
print('âœ“ OK')
```

### Ejercicio 12: Outer join
Outer merge

```python
# Pista: pd.merge(..., how='outer')
outer = pd.merge(cars1, cars2, how='outer')
```

```python
assert outer is not None
print('âœ“ OK')
```

### Ejercicio 13: On column
Merge on specific column

```python
# Pista: pd.merge(..., on='key')
on_col = pd.merge(cars1_clean, cars2, on='car')
```

```python
assert on_col is not None
print('âœ“ OK')
```

### Ejercicio 14: Suffixes
Add suffixes for conflicts

```python
# Pista: pd.merge(..., suffixes=('_left', '_right'))
suff = pd.merge(cars1_clean, cars2, on='car', suffixes=('_left', '_right'))
```

```python
assert suff is not None
print('âœ“ OK')
```

### Ejercicio 15: Append
Add rows with append/concat

```python
# Pista: pd.concat([cars1, cars2], ignore_index=True)
appended = pd.concat([cars1, cars2], ignore_index=True)
```

```python
assert len(appended) > 0
print('âœ“ OK')
```
