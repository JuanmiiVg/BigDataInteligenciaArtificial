### Ejercicio 7: Select 3 cols
Team, Yellow Cards, Red Cards

```python
# Pista: euro12[[...]]
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
```

```python
assert discipline.shape == (16, 3)
print('âœ“ OK')
```

### Ejercicio 8: Sort by cards
Red desc, Yellow desc

```python
# Pista: .sort_values()
sorted_d = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])
```

```python
assert sorted_d is not None
print('âœ“ Ordenado')
```

### Ejercicio 9: Media amarillas

```python
# Pista: .mean()
media_y = euro12['Yellow Cards'].mean()
```

### Ejercicio 10: Filter >6 goals
Teams with >6 goals

```python
# Pista: euro12[euro12.Goals > 6]
teams_6 = euro12[euro12['Goals'] > 6]
```

```python
assert len(teams_6) > 0
print('âœ“ OK')
```

### Ejercicio 11: Teams start G
Team name starts with G

```python
# Pista: .str.startswith('G')
teams_g = euro12[euro12['Team'].str.startswith('G')]
```

```python
assert len(teams_g) > 0
print('âœ“ OK')
```

### Ejercicio 12: First 7 cols
Select first 7

```python
# Pista: .iloc[:, 0:7]
first_7 = euro12.iloc[:, :7]
```

```python
assert first_7.shape[1] == 7
print('âœ“ OK')
```

### Ejercicio 13: No last 3
Exclude last 3

```python
# Pista: .iloc[:, :-3]
no_last = euro12.iloc[:, :-3]
```

```python
assert no_last.shape[1] == 32
print('âœ“ OK')
```

### Ejercicio 14: 3 teams accuracy
England, Italy, Russia

```python
# Pista: .loc[.isin(...)]
teams_p = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]
```

```python
assert teams_p.shape[0] == 3
print('âœ“ OK')
```

### Ejercicio 15: Complex filter
>5 goals AND >80 passes

```python
# Pista: (cond1) & (cond2)
teams_f = euro12[(euro12['Goals'] > 5) & (euro12['Passes'] > 80)]
```

```python
assert len(teams_f) > 0
print('âœ“ OK')
```
