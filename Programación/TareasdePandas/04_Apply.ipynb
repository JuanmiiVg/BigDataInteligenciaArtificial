### Ejercicio 4: Lambda upper
Create capitalize lambda

```python
# Pista: lambda x: x.capitalize()
cap = lambda x: x.capitalize()
```

```python
assert callable(cap)
print('âœ“ OK')
```

### Ejercicio 5: Apply capitalize
Capitalize Mjob, Fjob

```python
# Pista: .apply(capitalizer)
cap_cols = df[['Mjob', 'Fjob']].apply(lambda col: col.apply(cap))
```

```python
assert cap_cols is not None
print('âœ“ OK')
```

### Ejercicio 7: Apply inplace
Modify original df

```python
# Pista: df['Mjob'] = df['Mjob'].apply(...)
updated = df.copy()
updated['Mjob'] = updated['Mjob'].apply(cap)
updated['Fjob'] = updated['Fjob'].apply(cap)
```

```python
assert updated is not None
print('âœ“ OK')
```

### Ejercicio 8: Custom function
Create legal_drinker (age>17)

```python
# Pista: def majority(x): return x > 17
def legal(x):
    return x > 17
```

```python
assert callable(legal)
print('âœ“ OK')
```

### Ejercicio 9: Apply function
Add legal_drinker column

```python
# Pista: .apply(majority)
df_legal = df.copy()
df_legal['legal_drinker'] = df_legal['age'].apply(legal)
```

```python
assert df_legal is not None
print('âœ“ OK')
```

### Ejercicio 10: Lambda math
Multiply by 10

```python
# Pista: lambda x: x * 10 if isinstance(x, int) else x
mult = lambda x: x * 10 if isinstance(x, int) else x
```

```python
assert callable(mult)
print('âœ“ OK')
```

### Ejercicio 11: Applymap

```python
# Pista: .applymap() or .map()
all_mult = df.applymap(mult)
```

### Ejercicio 12: Groupby apply
Custom groupby function

```python
# Pista: .groupby('Mjob').apply(lambda x: x.mean())
group_apply = df.groupby('Mjob').apply(lambda x: x.mean(numeric_only=True))
group_apply
```

```python
assert group_apply is not None
print('âœ“ OK')
```

### Ejercicio 13: Transform
Transform with groupby

```python
# Pista: .groupby('school').transform('mean')
transformed = df.select_dtypes(include='number').groupby(df['school']).transform('mean')
```

```python
assert transformed is not None
print('âœ“ OK')
```

### Ejercicio 14: Agg multiple
Multiple aggregations

```python
# Pista: .agg(['mean', 'sum', 'std'])
multi_agg = df.select_dtypes(include='number').agg(['mean', 'sum', 'std'])
```

```python
assert multi_agg is not None
print('âœ“ OK')
```

### Ejercicio 15: Apply with axis
Apply by row/column

```python
# Pista: .apply(func, axis=1)
axis_apply = df.apply(lambda row: row['age'] + row['absences'], axis=1)
axis_apply
```

```python
assert axis_apply is not None
print('âœ“ OK')
```
