### Ejercicio 2: Create dict
Create Pokemon dictionary

```python
# Pista: data = {'Name': [...], 'Type': [...], ...}
pokemon_dict = {'Name': [], 'Type': [], 'HP': [], 'Evolution': [], 'Pokedex': []}
```

```python
assert isinstance(pokemon_dict, dict)
print('âœ“ OK')
```

### Ejercicio 3: Create DF
DataFrame from dict

```python
# Pista: pokemon = pd.DataFrame(pokemon_dict)
pokemon = pd.DataFrame(pokemon_dict)
```

```python
assert isinstance(pokemon, pd.DataFrame)
print('âœ“ OK')
```

### Ejercicio 4: Reorder cols
Reorder to: Name, Type, HP, Evolution, Pokedex

```python
# Pista: pokemon[['Name', 'Type', 'HP', 'Evolution', 'Pokedex']]
pokemon_ordered = pokemon[['Name', 'Type', 'HP', 'Evolution', 'Pokedex']]
```

```python
assert list(pokemon_ordered.columns)[0] == 'Name'
print('âœ“ OK')
```

### Ejercicio 5: Add column
Add 'Place' column

```python
# Pista: pokemon['Place'] = ['Kanto', 'Kanto', ...]
with_place = pokemon.copy()
with_place['Place'] = ['Kanto'] * len(pokemon)
```

```python
assert 'Place' in with_place.columns
print('âœ“ OK')
```

### Ejercicio 7: Create Series
Create a Series

```python
# Pista: pd.Series([1, 2, 3])
series = pd.Series([1, 2, 3])
```

```python
assert isinstance(series, pd.Series)
print('âœ“ OK')
```

### Ejercicio 8: Series with index
Series with custom index

```python
# Pista: pd.Series([...], index=['a', 'b', 'c'])
indexed = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
```

```python
assert len(indexed.index) == 3
print('âœ“ OK')
```

### Ejercicio 9: Concat series
Concatenate Series

```python
# Pista: pd.concat([s1, s2])
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])

concat_s = pd.concat([s1, s2])
```

```python
assert concat_s is not None
print('âœ“ OK')
```

### Ejercicio 10: From lists
DF from lists

```python
# Pista: pd.DataFrame({'col1': [...], 'col2': [...]})
df_lists = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6]})
```

```python
assert df_lists.shape[1] == 2
print('âœ“ OK')
```
