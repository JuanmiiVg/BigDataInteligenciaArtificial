# Visualization

### Dataset: Titanic

### Ejercicio 1: Import
Import pandas, matplotlib, seaborn

```python
# Pista: import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
assert 'pd' in dir()
print('✓ OK')
```

### Ejercicio 2: Load titanic
https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv

```python
# Pista: pd.read_csv()
titanic = pd.read_csv(
    "train.csv"
)
```

```python
assert titanic.shape[0] > 0
print('✓ Cargado')
```

### Ejercicio 3: Set index
PassengerId as index

```python
# Pista: .set_index('PassengerId')
titanic_idx = titanic.set_index('PassengerId')
titanic_idx
```

```python
assert 'PassengerId' in titanic_idx.index.name or titanic_idx.index.name is None
print('✓ OK')
```

### Ejercicio 4: Pie gender
Gender distribution pie

```python
# Pista: titanic['Sex'].value_counts().plot(kind='pie')
pie = titanic['Sex'].value_counts().plot(kind='pie')
```

```python
assert pie is not None
print('✓ OK')
```

### Ejercicio 5: Scatter
Fare vs Age by gender

```python
# Pista: .plot(x='Age', y='Fare', kind='scatter', c='Sex')
scatter = titanic.plot(
    x='Age',
    y='Fare',
    kind='scatter',
    c=titanic['Sex'].map({'male': 0, 'female': 1}),
    cmap='coolwarm'
)
```

```python
assert scatter is not None
print('✓ OK')
```

### Ejercicio 6: Survived count
How many survived?

```python
# Pista: titanic['Survived'].sum()
survived = titanic['Survived'].sum()
survived
```

```python
assert survived > 0
print('✓ OK')
```

### Ejercicio 7: Histogram fare
Fare distribution

```python
# Pista: titanic['Fare'].plot(kind='hist')
hist = titanic['Fare'].plot(kind='hist')
hist
```

```python
assert hist is not None
print('✓ OK')
```

### Ejercicio 8: Bar plot
Bar chart of classes

```python
# Pista: titanic['Pclass'].value_counts().plot(kind='bar')
bar = titanic['Pclass'].value_counts().plot(kind='bar')
bar
```

```python
assert bar is not None
print('✓ OK')
```

### Ejercicio 9: Boxplot age
Age boxplot

```python
# Pista: titanic['Age'].plot(kind='box')
box = titanic['Age'].plot(kind='box')
box
```

```python
assert box is not None
print('✓ OK')
```

### Ejercicio 10: Group viz
Survived by gender

```python
# Pista: titanic.groupby('Sex')['Survived'].mean()
surv_gender = titanic.groupby('Sex')['Survived'].mean()
surv_gender
```

```python
assert surv_gender is not None
print('✓ OK')
```

### Ejercicio 11: Heatmap
Correlation heatmap

```python
# Pista: sns.heatmap(titanic.corr())
heat = sns.heatmap(titanic.select_dtypes(include='number').corr())
plt.show()
```

```python
assert heat is not None
print('✓ OK')
```

### Ejercicio 12: Kde plot
Age KDE

```python
# Pista: titanic['Age'].plot(kind='kde')
kde = titanic['Age'].plot(kind='kde')
```

```python
assert kde is not None
print('✓ OK')
```

### Ejercicio 13: Multi plot
Age by Pclass

```python
# Pista: titanic.boxplot(column='Age', by='Pclass')
multi = titanic.boxplot(column='Age', by='Pclass')
```

```python
assert multi is not None
print('✓ OK')
```

### Ejercicio 14: Violin
Violin plot

```python
# Pista: sns.violinplot(data=titanic, x='Sex', y='Age')
violin = sns.violinplot(data=titanic, x='Sex', y='Age')
```

```python
assert violin is not None
print('✓ OK')
```
