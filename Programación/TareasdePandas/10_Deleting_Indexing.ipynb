### Ejercicio 1: Import
Import pandas, numpy

```python
# Pista: import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
assert 'pd' in dir()
print('âœ“ OK')
```

### Ejercicio 2: Load iris
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

```python
# Pista: pd.read_csv()
iris = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
    header=None
)
```

```python
assert iris.shape[0] > 0
print('âœ“ Cargado')
```

### Ejercicio 3: Add columns
sepal_length, sepal_width, petal_length, petal_width, class

```python
# Pista: iris.columns = [...]
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_cols = iris
```

```python
assert iris_cols.shape[1] == 5
print('âœ“ OK')
```

### Ejercicio 4: Check nulls
Missing values?

```python
# Pista: .isnull().sum()
nulls = iris.isnull().sum()
```

```python
assert nulls is not None
print('âœ“ OK')
```

### Ejercicio 5: Set NaN
petal_length rows 10-29 to NaN

```python
# Pista: iris.loc[10:29, 'petal_length'] = np.nan
iris_nan = iris.copy()
iris_nan.loc[10:29, 'petal_length'] = np.nan
```

```python
assert iris_nan.isnull().sum().sum() > 0
print('âœ“ OK')
```

### Ejercicio 6: Fill NaN
Fill NaN with 1.0

```python
# Pista: .fillna(1.0)
iris_fill = iris_nan.fillna(1.0)
```

```python
assert iris_fill.isnull().sum().sum() == 0
print('âœ“ OK')
```

### Ejercicio 7: Drop class
Delete class column

```python
# Pista: .drop('class', axis=1)
iris_no_class = iris.drop('class', axis=1)
```

```python
assert 'class' not in iris_no_class.columns
print('âœ“ OK')
```

### Ejercicio 8: First 3 NaN
Set first 3 rows to NaN

```python
# Pista: iris.loc[0:2] = np.nan
iris_head_nan = iris.copy()
iris_head_nan.loc[0:2] = np.nan
```

```python
assert iris_head_nan.isnull().any().any()
print('âœ“ OK')
```

### Ejercicio 9: Dropna
Delete rows with NaN

```python
# Pista: .dropna()
iris_clean = iris.dropna()
```

```python
assert iris_clean.isnull().sum().sum() == 0
print('âœ“ OK')
```

### Ejercicio 10: Reset index
Reset index to 0,1,2,...

```python
# Pista: .reset_index(drop=True)
iris_reset = iris_clean.reset_index(drop=True)
```

```python
assert iris_reset.index[0] == 0
print('âœ“ OK')
```

### Ejercicio 11: Set index
Set column as index

```python
# Pista: .set_index('sepal_length')
iris_indexed = iris.set_index('sepal_length')
```

```python
assert iris_indexed.index.name == 'sepal_length'
print('âœ“ OK')
```

### Ejercicio 12: Drop duplicates
Remove duplicate rows

```python
# Pista: .drop_duplicates()
iris_unique = iris.drop_duplicates()
```

```python
assert iris_unique is not None
print('âœ“ OK')
```

### Ejercicio 13: Filter index
Select by index value

```python
# Pista: .loc[5.0]
by_index = iris_indexed.loc[5.0]
```

```python
assert by_index is not None
print('âœ“ OK')
```

### Ejercicio 14: Multi index
Create MultiIndex

```python
# Pista: pd.MultiIndex.from_tuples([...])
multi_idx = pd.MultiIndex.from_tuples([
    ('A', 1),
    ('A', 2),
    ('B', 1),
    ('B', 2)
])
```

```python
assert multi_idx is not None
print('âœ“ OK')
```
