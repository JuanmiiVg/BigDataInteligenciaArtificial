## 游닇 Ejercicio 5: Diccionarios y listas

```python
# Crea una lista con 3 diccionarios que representen alumnos (nombre y nota)
# Recorre la lista e imprime el nombre del alumno y si est치 aprobado (nota >= 5)

# Tu c칩digo aqu칤

alumnos = [
    {"nombre": "Laura", "nota": 7},
    {"nombre": "Juan", "nota": 4},
    {"nombre": "Manuel", "nota": 9}
]

for alumno in alumnos:
    nombre = alumno["nombre"]
    nota = alumno["nota"]
    estado = "Aprobado" if nota >= 5 else "Suspenso"
    print(f"{nombre} - {estado}")

```
