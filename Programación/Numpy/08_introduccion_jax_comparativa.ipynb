## 2. Arrays y operaciones b谩sicas con `jnp`

```
try:
    x = jnp.array([1.0, 2.0, 3.0])
    y = jnp.array([0.5, 0.5, 0.5])
    print(x + y)
    print(jnp.sqrt(x))
except NameError:
    print("Salta esta celda si JAX no est谩 disponible.")
```

## 3. `jit`: compilaci贸n just-in-time

```
try:
    import time
    def f_np(a, b):
        return a * a + b

    import numpy as np
    a_np = np.random.rand(1_000_00)
    b_np = np.random.rand(1_000_00)

    # NumPy puro
    t0 = time.time()
    out_np = f_np(a_np, b_np)
    t1 = time.time()

    # JAX (primera vez compila, segunda es m谩s r谩pida)
    import jax, jax.numpy as jnp
    @jax.jit
    def f_jax(a, b):
        return a * a + b

    a_j = jnp.array(a_np)
    b_j = jnp.array(b_np)
    t2 = time.time()
    out_j1 = f_jax(a_j, b_j).block_until_ready()
    t3 = time.time()
    out_j2 = f_jax(a_j, b_j).block_until_ready()
    t4 = time.time()

    print(f"NumPy: {t1 - t0:.4f}s")
    print(f"JAX (1陋 con compilaci贸n): {t3 - t2:.4f}s")
    print(f"JAX (2陋 ya compilado):    {t4 - t3:.4f}s")
    print("Igualdad aproximada:", np.allclose(np.array(out_j2), out_np))
except Exception as e:
    print("No se pudo ejecutar JAX jit:", e)
```

## 4. `grad`: derivadas autom谩ticas

```
try:
    import jax.numpy as jnp, jax
    def f(x):
        return jnp.sin(x) + x**2

    df = jax.grad(f)
    print("f'(2.0) ", df(2.0))
except Exception as e:
    print("No se pudo usar grad:", e)
```

## 5. `vmap`: vectorizaci贸n de funciones

```
try:
    import jax.numpy as jnp, jax
    def escala(x, s):
        return s * x

    escala_v = jax.vmap(escala, in_axes=(0, None))
    xs = jnp.arange(5.0)
    print(escala_v(xs, 2.0))  # escala cada elemento por 2
except Exception as e:
    print("No se pudo usar vmap:", e)
```
