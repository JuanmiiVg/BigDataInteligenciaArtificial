## 1锔 Ejercicio 1
**Datos:** 10, 13, 6, 28, 31

Calcular:
- Desviaci贸n media
- Varianza
- Desviaci贸n t铆pica

```python
import numpy as np

# Datos
datos1 = np.array([10, 13, 6, 28, 31])

# Media
media1 = np.mean(datos1)

# Desviaci贸n media
desviacion_media1 = np.mean(np.abs(datos1 - media1))

# Varianza
varianza1 = np.var(datos1, ddof=0)

# Desviaci贸n t铆pica
desviacion_tipica1 = np.sqrt(varianza1)

print(f"Datos: {datos1}")
print(f"Media: {media1:.2f}")
print(f"Desviaci贸n media: {desviacion_media1:.2f}")
print(f"Varianza: {varianza1:.2f}")
print(f"Desviaci贸n t铆pica: {desviacion_tipica1:.2f}")
```

## 2锔 Ejercicio 2
Calcular los mismos valores para las siguientes series:
- Serie A: 2, 3, 6, 8, 11
- Serie B: 12, 6, 7, 3, 15, 10, 18, 5

```python
# Serie A
serieA = np.array([2, 3, 6, 8, 11])
mediaA = np.mean(serieA)
desv_mediaA = np.mean(np.abs(serieA - mediaA))
varA = np.var(serieA)
desv_tipA = np.sqrt(varA)

# Serie B
serieB = np.array([12, 6, 7, 3, 15, 10, 18, 5])
mediaB = np.mean(serieB)
desv_mediaB = np.mean(np.abs(serieB - mediaB))
varB = np.var(serieB)
desv_tipB = np.sqrt(varB)

print("Serie A:")
print(f"  Media = {mediaA:.2f}")
print(f"  Desviaci贸n media = {desv_mediaA:.2f}")
print(f"  Varianza = {varA:.2f}")
print(f"  Desviaci贸n t铆pica = {desv_tipA:.2f}\n")

print("Serie B:")
print(f"  Media = {mediaB:.2f}")
print(f"  Desviaci贸n media = {desv_mediaB:.2f}")
print(f"  Varianza = {varB:.2f}")
print(f"  Desviaci贸n t铆pica = {desv_tipB:.2f}")
```

## 3锔 Ejercicio 3
**Datos:** 3, 8, 4, 10, 6, 2

Calcular:
- Media y varianza
- Si todos los datos se multiplican por 3, 驴cu谩l ser铆a la nueva media y varianza?

```python
# Datos originales
datos3 = np.array([3, 8, 4, 10, 6, 2])

media3 = np.mean(datos3)
varianza3 = np.var(datos3)

# Multiplicamos todos los datos por 3
datos3_nuevo = datos3 * 3
media3_nueva = np.mean(datos3_nuevo)
varianza3_nueva = np.var(datos3_nuevo)

print(f"Datos originales: {datos3}")
print(f"Media original: {media3:.2f}")
print(f"Varianza original: {varianza3:.2f}\n")

print(f"Datos multiplicados por 3: {datos3_nuevo}")
print(f"Nueva media: {media3_nueva:.2f}")
print(f"Nueva varianza: {varianza3_nueva:.2f}")
```

###  Observaci贸n te贸rica
Cuando multiplicamos todos los datos por una constante **k**:
- La **media** se multiplica por **k**.
- La **varianza** se multiplica por **k虏**.
- La **desviaci贸n t铆pica** se multiplica por **|k|**.
